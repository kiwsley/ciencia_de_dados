```python
#VisualizaÃ§Ã£o
from datetime import datetime, timedelta
base["Order Date"]=pd.to_datetime(base["Order Date"], dayfirst=True)
base["Ship Date"]=pd.to_datetime(base["Ship Date"], dayfirst=True)
```

```python
base.info()
```

```python
base["Ano"]=base["Order Date"].dt.year
```

```python
base.head()
```

```python
base.groupby("Ano")["Sales"].sum().plot.bar();
```

```python
import matplotlib.pyplot as plt
import numpy as np

index= base.groupby("Ano")["Sales"].sum().index
valor= base.groupby("Ano")["Sales"].sum().values

fig, ax = plt.subplots(figsize=(8,5))

ax.bar(index,valor, linewidth=2.0)
ax.set_title("Soma das Vendas dos Produtos por Ano", fontsize = 15, fontweight="bold")

ax.xaxis.set_ticks([2015,2016,2017,2018])
ax.tick_params(axis='x',labelsize=12)

for i in range(0,4):
    ax.annotate("{:.2f}".format(valor[i]), #texto
                (index[i],valor[i]), #posiÃ§Ã£o,
                ha='center',
                )
ax.yaxis.set_visible(False)    
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)


                

plt.show()
```

```python
base["Mes"]= base['Order Date'].dt.month
base.head()
```

```python
soma_mes =base.groupby(["Mes","Ano"])["Sales"].sum()
soma_mes.head()
```

```python
soma_mes=soma_mes.reset_index()
```

```python
soma_mes.head()
```

```python
soma_mes[soma_mes.Ano ==2015].Mes.values
```

```python
fig, ax = plt.subplots(figsize=(16,8))

wid= 0.2

Ano2015 = soma_mes[soma_mes.Ano ==2015].Mes.values-wid-wid/2
vendas2015= soma_mes[soma_mes.Ano ==2015].Sales.values

Ano2016 = soma_mes[soma_mes.Ano ==2016].Mes.values-wid/2
vendas2016= soma_mes[soma_mes.Ano ==2016].Sales.values

Ano2017 = soma_mes[soma_mes.Ano ==2017].Mes.values+wid/2
vendas2017= soma_mes[soma_mes.Ano ==2017].Sales.values

Ano2018 = soma_mes[soma_mes.Ano ==2018].Mes.values+wid+wid/2
vendas2018= soma_mes[soma_mes.Ano ==2018].Sales.values

ax.bar(Ano2015,vendas2015, linewidth=2.0,width=0.2,label=2015)
ax.bar(Ano2016,vendas2016, linewidth=2.0,width=0.2,label=2016)
ax.bar(Ano2017,vendas2017, linewidth=2.0,width=0.2,label=2017)
ax.bar(Ano2018,vendas2018, linewidth=2.0,width=0.2,label=2018)
ax.set_title("Soma das Vendas dos Produtos por Ano", fontsize = 20, fontweight="bold")

for i in range(0,12):
    ax.annotate("{:.2f}".format(vendas2018[i]/1000), #texto
                (Ano2018[i],vendas2018[i]), #posiÃ§Ã£o,
                ha='center',
                va="top",
                xytext=(0,+30),#descolamnento do texto
                textcoords="offset points",
                fontsize = 16,
                color='red',
                fontweight='bold'
                )
    

ax.yaxis.set_visible(False)    
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks(np.arange(1,13))
ax.tick_params(axis='x',labelsize=15)
ax.legend()
```

```python
#Qual o item mais vendido? 

itens =base.groupby("Product Name")["Sales"].sum()
itens = itens.reset_index()
itens= itens.sort_values("Sales",ascending=False)
itens.head(10)
```

```python
n=5
top_n = itens.head(n)
top_n

top_n["Product Name"].values
```

```python
fig, ax = plt.subplots(figsize=(16,10))

ax.barh(top_n["Product Name"].values,
        top_n["Sales"].values)

ax.invert_yaxis() #invetendo a ordem
ax.set_title("Top n Itens mais vendidos",fontsize = 20, fontweight="bold")

for i in range(0,n):
    ax.annotate("{:.2f}".format(top_n["Sales"].values[i]/1000), #texto
                (top_n["Sales"].values[i],i), #posiÃ§Ã£o,
                ha='center',
                va="top",
                xytext=(30,+2),#descolamnento do texto
                textcoords="offset points",
                fontsize = 16,
                color='red',
                fontweight='bold'
                )
    
ax.yaxis.set_visible(False)    
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
#ax.xaxis.set_ticks(np.arange(2015,2019))
ax.tick_params(axis='x',labelsize=15)

plt.show()
```

```python
#agrupamento item por ano

itens_ano = base.groupby(["Product Name","Ano"]).Sales.sum()

itens_ano= itens_ano.reset_index()
itens_ano.head(10)
```

```python
#juntando base no Pandas

dt_datas = pd.DataFrame([2018,2017,2016,2015])
dt_datas.columns=['Ano']

dt_datas
```

```python
dt_datas['uniao']='unir'
itens['uniao']='unir'

dt_datas
```

```python
produto_ano= pd.merge(
    itens,
    dt_datas,
    how="outer",
    on='uniao'
)

produto_ano.head(10)
```

```python
itens_ano_grafico= pd.merge(
    produto_ano,
    itens_ano,
    how="left",
    on=["Product Name",'Ano']
)

itens_ano_grafico.head(10)
```

```python
itens_ano_grafico[itens_ano_grafico.Ano ==2018]["Product Name"]
```

```python
itens_ano_grafico[itens_ano_grafico.Ano == 2018].Sales_y.values
```

```python
fig, ax = plt.subplots(figsize=(16,12))

qtde_itens = len(top_n)

ax.barh(np.arange(0,qtde_itens),
        itens_ano_grafico[itens_ano_grafico.Ano == 2018].Sales_y.values,
        align="center")

plt.show()
```
