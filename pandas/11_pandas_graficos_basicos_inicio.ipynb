# Pandas b√°sico

```python
import pandas as pd

BASE = "dados/cereal_sem_negativos.csv"

df = pd.read_csv(BASE)

df.head()
```

```python
df["mfr"].value_counts().plot(kind="bar")
```

```python
df["mfr"].value_counts().plot(
    kind="bar",
    xlabel= "Fabricante/Empresas",
    ylabel="quantidade",
    title= "Contagem de Cereais Por Fabricante")
```

```python
df["mfr"].value_counts().plot(
    kind="pie",
    ylabel="Quantidade",
    autopct ="%.1f%%",
    title= "Contagem de Cereais Por Fabricante")
```

```python
df.groupby("mfr")["calories"].mean().plot(
    kind="bar",
    xlabel="Fabricante",
    ylabel="Calorias",
    title="M√©dia de calorias por fabricante"
)
```
