```python
serie_dolar = "DEXBZUS"

URL_serie_dolar =f"https://api.stlouisfed.org/fred/series/observations?series_id={serie_dolar}&api_key={key}&file_type=json"

reposta_dolar = requests.get(URL_serie_dolar)
reposta_dolar.status_code
```

```python
df_dollar = pd.DataFrame(reposta_dolar.json()["observations"])

df_dollar
```

```python
df_dollar.info()
```

```python
df_dollar=df_dollar.drop(columns=["realtime_start","realtime_end"])

df_dollar["date"]=pd.to_datetime(df_dollar["date"])

df_dollar.info()
```

```python
df_dollar["value"]=df_dollar["value"].replace(".",float("NaN"))
df_dollar["value"]=df_dollar["value"].astype(float)
```

```python
df_dollar=df_dollar.set_index("date")
df_dollar.head()
```

```python
df_dollar.plot(figsize=(14,8))
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(14,6))

ax.plot(df_dollar.index,df_dollar["value"])
ax.set_title("Cotação do Dólar no Brasil")
ax.set_ylabel("Cotação do Dólar")
ax.set_xlabel("Data")

ax.grid(linestyle="--", linewidth=0.5, color="black",alpha =0.3)

plt.show()
```

```python
presidentes={
    "FHC":("1995-01-01","2002-12-31"),
    "LULA 1 2":("2003-01-01","2010-12-31"),
    "DILMA":("2011-01-01","2016-08-31"),
    "TEMER":("2016-09-01","2018-12-31"),
    "BOLSONARO":("2019-01-01","2022-12-31"),
    "LULA 3":("2023-01-01", "2025-01-09"),
}
```

```python
fig, ax = plt.subplots(figsize=(14,6))

ax.plot(df_dollar.index,df_dollar["value"])
ax.set_title("Cotação do Dólar no Brasil")
ax.set_ylabel("Cotação do Dólar")
ax.set_xlabel("Data")

ax.grid(linestyle="--", linewidth=0.5, color="black",alpha =0.3)

cores=plt.cm.Pastel2.colors

for presidente, periodo, cor in zip(presidentes.keys(),presidentes.values(),cores):
    ax.axvspan(periodo[0],periodo[1], color=cor,alpha=0.5,label=presidente)

from matplotlib.dates import YearLocator

ax.xaxis.set_major_locator(YearLocator(2))
ax.xaxis.set_tick_params(rotation = 45)


ax.legend()
plt.show()
```

```python
fig, axs = plt.subplots(nrows=3,ncols=2,figsize=(12,10),tight_layout = True)



for presidente, periodo, ax, in zip(presidentes.keys(),presidentes.values(), axs.flatten()):
    df_temp = df_dollar.loc[presidentes[presidente][0]:presidentes[presidente][1]]
    ax.xaxis.set_tick_params(rotation = 45)

    ax.plot(df_temp.index, df_temp["value"])
    ax.set_title(f"{presidente} ({periodo[0]} - ({periodo[1]}))")

fig.suptitle("Cotação do Dolar no Brasil - Plano Real")
plt.show()
```
