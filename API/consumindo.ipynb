```python
import requests

URL_DDD ="https://brasilapi.com.br/api/ddd/v1/"

resposta_ddd_21 = requests.get(URL_DDD + "21" )
resposta_ddd_21 = requests.get("".join([URL_DDD,"21"]))
resposta_ddd_21.status_code
```

```python
resposta_ddd_21.json()
```

```python
import pandas as pd

df_ddd_21 = pd.DataFrame(resposta_ddd_21.json()).assign(ddd=21)

df_ddd_21
```

```python
for ddd in range(20,31):
    resposta = requests.get("".join([URL_DDD,str(ddd)]))
    print(f"DDD:{ddd} - Status code:{resposta.status_code}")
```

```python
df_DDD = pd.DataFrame()

for ddd in range(20,31):
    resposta = requests.get("".join([URL_DDD,str(ddd)]))
    if resposta.status_code == 200:
        df_DDD = pd.concat([df_DDD, pd.DataFrame(resposta.json()).assign(ddd=ddd)])

df_DDD.reset_index(drop=True)
```
