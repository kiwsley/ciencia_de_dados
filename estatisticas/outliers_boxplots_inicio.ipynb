### Medidas de posição ou localização

As medidas de posição ou localização são utilizadas para descrever a posição dos dados em relação ao eixo dos valores assumidos pela variável. São subdivididas em medidas de tendência central (média, mediana e moda) e medidas separatrizes (quartis, decis e percentis).

#### Medidas de tendência central

A moda (*Mo*) é o valor que ocorre com maior frequência em um conjunto de dados. É a única medida de tendência central que pode ser utilizada para variáveis qualitativas, já que essas variáveis permitem a identificação de valores repetidos.

Para dados não agrupados, a moda é o valor que ocorre com maior frequência. Para dados contínuos agrupados, existem procedimentos para estimar a moda, como a moda de King e a moda de Czuber.

A média aritmética pode ser a medida representativa de uma população com *N* elementos, representada pela letra grega *μ* (mi), ou a medida representativa de uma amostra com *n* elementos, representada por $\bar{X}$ (x barra).

A forma de calcula depende se a variável é discreta, contínua, agrupada ou não agrupada. Além disso, a média pode ter pesos iguais ou diferentes para cada observação, dando origem à média aritmética simples e à média ponderada, respectivamente.

Considerando dados não agrupados, a média aritmética simples é calculada pela fórmula:

$$
\bar{X} = \frac{\sum\limits_{i=1}^{n} X_{i}}{n}
$$

em que *n* é o número de observações e *X<sub>i</sub>* é o valor da variável para cada observação.

```python
# confirmando a média
df_cenouras["carrots"].sum() / df_cenouras["carrots"].count()
```

A mediana (*Md*) é uma medida de tendência central que divide o conjunto de dados em duas partes iguais, demo que 50% dos elementos são menores ou iguais à mediana e os outros 50% são maiores ou iguais à mediana. Para calcular a mediana, os dados devem ser ordenados em ordem crescente. 

Para dados não agrupados, a mediana é calculada pela fórmula:

$$
Md(X) = \left\{ \begin{array}{ll}
X_{\left(\frac{n+1}{2}\right)} & \text{se } n \text{ ímpar} \\
\frac{X_{\left(\frac{n}{2}\right)} + X_{\left(\frac{n}{2}+1\right)}}{2} & \text{se } n \text{ par}
\end{array} \right.
$$

em que *n* é o número de observações e $X_i \leq \ldots \leq X_n$, tal que *X<sub>(i)</sub>* é a menor observação ou o valor do primeiro elemento e *X<sub>(n)</sub>* é a maior observação ou o valor do último elemento.

#### Medidas separatrizes

O uso apenas de medidas de tendência central pode não ser suficiente para representar um conjunto de dados, já que estes podem ser afetados por valores extremos. Além disso, a dispersão e simetria dos dados também são importantes para a descrição do conjunto.

As medidas separatrizes são utilizadas para dividir um conjunto de dados em partes iguais. Os quartis, decis e percentis são exemplos de medidas separatrizes.

Na realidade, tais medidas são casos específicos de quantis, os quais são valores que dividem um conjunto de dados ordenado em partes iguais.

Os quartis são medidas que dividem um conjunto de dados ordenado crescentemente em quatro partes iguais. 

<br>



Assim, o 1º Quartil (Q<sub>1</sub> ou 25° percentil) indica que 25% dos dados são inferiores a Q<sub>1</sub> ou que 75% dos dados são superiores a Q<sub>1</sub>.

O 2º Quartil (Q<sub>2</sub> ou 5° decil ou 50° percentil) corresponde à mediana, indicando que 50% dos dados são inferiores ou superiores a Q<sub>2</sub>.

Já o 3º Quartil (Q<sub>3</sub> ou 75° percentil) indica que 75% dos dados são inferiores a Q<sub>3</sub> ou que 25% dos dados são superiores a Q<sub>3</sub>.

Os decis são medidas que dividem um conjunto de dados ordenados crescentemente em 10 partes iguais.

<br>



A lógica é a mesma dos quartis no que diz respeito à intrepretação dos valores. O 1º Decil (D<sub>1</sub> ou 10° percentil) indica que 10% dos dados são inferiores a D<sub>1</sub> ou que 90% dos dados são superiores a D<sub>1</sub>.

Os percentis são medidas que dividem um conjunto de dados ordenados crescentemente em 100 partes iguais. Assim, o 1° percentil indica que 1% dos dados são inferiores a ele, o 2° percentil indica que 2% dos dados são inferiores a ele, e assim por diante.

#### Identificação de valores extremos (outliers) univariados

Um conjunto de dados pode conter algumas observações que apresentam grande afastamento das demais. Essas observações são chamadas de valores extremos, valores atípicos, ou outliers.

O que fazer com esses valores? A decisão de manter ou excluir os outliers depende do objetivo da análise e da origem de tais valores. Em geral, os outliers podem ser mantidos, excluídos ou tratados.

Existem vários métodos de identificação de outliers: boxplot, modelos de discordância, teste de Dixon, teste de Grubbs, Zscores, entre outros.


Um boxplot é um método para demonstrar graficamente a localidade (tendência central), dispersão e assimetria de grupos de dados numéricos através de seus quartis. 

Além da caixa em um boxplot, pode haver linhas (chamadas de *whiskers*, bigodes em portuês) que se estendem da caixa indicando variabilidade (distância entre os valores) fora dos quartis superiores e inferiores. Outliers que diferem significativamente do restante do conjunto de dados podem ser plotados como pontos individuais além dos bigodes no boxplot. Boxplots são não-paramétricos: eles exibem variação em amostras de uma população estatística sem fazer nenhuma suposição sobre a distribuição estatística subjacente. Os espaçamentos em cada subseção do boxplot indicam o grau de dispersão e assimetria dos dados. Boxplots podem ser desenhados horizontalmente ou verticalmente.

Um boxplot é uma maneira padronizada de exibir o conjunto de dados com base no resumo dos cinco números: o mínimo, o máximo, a mediana da amostra e o primeiro e terceiro quartis.

- Mínimo (Q0 ou 0º percentil): o ponto de dados mais baixo no conjunto de dados, excluindo quaisquer outliers
- Máximo (Q4 ou 100º percentil): o ponto de dados mais alto no conjunto de dados, excluindo quaisquer outliers
- Mediana (Q2 ou 50º percentil): o valor do meio no conjunto de dados
- Primeiro quartil (Q1 ou 25º percentil): também conhecido como quartil inferior, é a mediana da metade inferior do conjunto de dados.
- Terceiro quartil (Q3 ou 75º percentil): também conhecido como quartil superior, é a mediana da metade superior do conjunto de dados.

Além dos valores mínimo e máximo usados para construir um boxplot, outro elemento importante que também pode ser empregado para obter um boxplot é a amplitude interquartil (IQR), a distância entre os quartis superior e inferior.

Para comparar dois ou mais boxplots:

- Compare as medianas. Se a linha da mediana de um boxplot está fora da caixa de outro boxplot, então os dois grupos comparados são provavelmente diferentes.
- Compare os IQRs e os whiskers para ter uma ideia das diferentes dispersões.
- Compare o comportamento dos outliers.
- Compare *skewness* (assimetria)

```python
import matplotlib.pyplot as plt

fig, (ax1,ax2) = plt.subplots(
    nrows=2,
    ncols=1,
    sharex=True,
    gridspec_kw={
        "height_ratios":(0.15,0.85),
        "hspace":0.02,
        }
        )



sns.boxplot(data = df_cenouras, 
            x="carrots" , 
            showmeans=True,
            ax =ax1,
            meanline=True,
            meanprops={"color":"C1","linewidth":1.5},
            medianprops={"color":"C2","linewidth":1.5}
            )

sns.histplot(data=df_cenouras,x="carrots", kde=True, bins="sturges",ax = ax2)

for ax in (ax1,ax2):
    ax.grid(True,linestyle='--',color = 'gray', alpha=0.2)
    ax.set_axisbelow(True)

ax2.axvline(df_cenouras['carrots'].mean(), color ="C1", linestyle ="--", label="Média")
ax2.axvline(df_cenouras['carrots'].median(), color ="C2", linestyle ="--", label="Mediana")
ax2.axvline(df_cenouras['carrots'].mode()[0], color ="C3", linestyle ="--", label="Moda")


ax2.legend()

plt.show()
```
