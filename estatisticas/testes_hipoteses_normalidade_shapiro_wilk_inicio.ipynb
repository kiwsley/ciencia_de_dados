# Testes para normalidade univariada

## Teste de Shapiro-Wilk

Teste específico para distribuições normais e pode ser aplicado para amostras de tamanho $4 \le n \le 5000$, sendo uma alternativa ao teste K-S para normalidade de pequenas amostras.

O teste de Shapiro-Wilk assume as seguintes hipóteses:

- $H_0$: a amostra provém de uma população com distribuição normal
- $H_1$: a amostra não provém de uma população com distribuição normal

Exemplo: Considere uma empresa que produz aviões. A base a seguir apresenta os dados de produção mensal de aviões nos últimos 24 meses. Verifique se os dados são provenientes de uma população com distribuição normal, considerando $\alpha = 1 \%$.

```python
import pandas as pd

DADOS_PRODUCAO_AVIOES = "./dados/aircraft.csv"

df_avioes = pd.read_csv(DADOS_PRODUCAO_AVIOES)
df_avioes.head()
```

```python
from scipy.stats import shapiro

shapiro(df_avioes["production"])
```
