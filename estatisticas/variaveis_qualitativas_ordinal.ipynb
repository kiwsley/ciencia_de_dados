```python
classificacao_produto = [
    "pessimo",
    "otimo",
    "otimo",
    "ruim",
    "regular",
    "bom",
    "bom",
    "bom",
    "ruim",
    "pessimo",
    "regular",
    "bom",
    "otimo",
    "pessimo",
    "bom",
    "bom",
    "regular",
    "otimo",
    "otimo",
    "regular",
]

len(classificacao_produto)
```

```python
import pandas as pd
df_classificacao = pd.DataFrame(classificacao_produto, columns =['classificacao'] )
```

```python
df_classificacao.info()
```

```python
df_classificacao.describe()
```

```python
df_classificacao['classificacao'].value_counts()
```

```python
df_classificacao['classificacao'].value_counts().sort_index()
```

```python
categorias =['pessimo','ruim','regular','bom','otimo']

from pandas.api.types import CategoricalDtype

cat_type = CategoricalDtype(categories=categorias,ordered=True)

df_classificacao['classificacao']=df_classificacao['classificacao'].astype(cat_type)

df_classificacao.info()
```

```python
df_classificacao['classificacao'].value_counts().sort_index()
```

```python
df_classificacao['classificacao'].value_counts().plot(kind='bar')
```
