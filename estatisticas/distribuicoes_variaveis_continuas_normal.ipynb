# Distribuições de probabilidade para variáveis contínuas

## Distribuição normal

A **distribuição normal**, também conhecida como **distribuição Gaussiana**, é a distribuição de probabilidade mais utilizada e importante, pois permite modelar uma infinidade de fenômenos naturais, estudos do comportamento humano, processos industriais, entre outros, além de possibilitar o uso de aproximações para o cálculo de probabilidades de muitas variáveis aleatórias.

Uma variável alatória $X$ com média $\mu \in \mathbb{R}$ e desvio-padrão $\sigma > 0$ tem distribuição normal se sua função densidade de probabilidade for dada por:

$$
f(x) = \frac{1}{\sigma \sqrt{2 \pi}} \exp \left( -\frac{(x - \mu)^2}{2 \sigma^2} \right) \qquad -\infty \le x \le +\infty
$$

Há a chamada **distribuição normal padrão**, a variável original é transformada em uma nova variável aleatória $Z$, com média zero ($\mu = 0$) e variância 1 ($\sigma^2 = 1$):

$$
Z = \frac{X - \mu}{\sigma}
$$

O score Z representa o número de desvios-padrão que separa uma variável aleatória X da média.

Este tipo de transformação, conhecida por *Zscores*, é muito utilizada para a padronização de variáveis, pois não altera a forma da distribuição da variável original e gera uma nova variável com média zero e variância 1. Desta forma, quando muitas variáveis com diferentes métricas ou ordens de grandeza estiverem sendo utilizadas em determinada modelagem, o processo de padronização *Zscores* fará com que todas as novas variáveis padronizadas apresentem média zero e variância 1.

Para amostras, podemos escrever

$$
Z = \frac{X - \bar{x}}{S}
$$

Com a variável aleatória $Z$:

$$
f(z) = \frac{1}{\sqrt{2\pi}} \exp \left( - \frac{z^2}{2} \right)
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm

normal = norm.rvs(0 ,1, size=10_000, random_state=42)

valores, contagem = np.unique(normal,return_counts=True)
print(valores)
print(contagem)

plt.scatter(valores, contagem,marker='o',color='red')

plt.show()
```

```python
x = np.linspace(-5,5,10_000)
y = norm.pdf(x,0,1)

plt.scatter(x,y,marker ='o', color='red')
plt.plot(x,y)

plt.show()
```

```python
import seaborn as sns
fig, ax = plt.subplots()

sns.histplot(normal, ax=ax,stat='density'
             )

ax.plot(x,y)

plt.show()
```

```python
x = np.linspace(-4,4,10_000)
y = norm.cdf(x,0,1)

plt.scatter(x,y,marker ='o', color='red')
plt.plot(x,y)

plt.show()
```
