# Distribuições de probabilidade para variáveis discretas

Há diversas distribuições existentes. Aqui, estudaremos algumas mais recorrentes.

## Uniforme

Todos os valores têm a mesma probabilidade de ocorrência:

$$ P(X=x_i) = \frac{1}{n} $$

onde $n$ representa a quantidade de possíveis valores, $X$ é a variável aleatória representando o valor e $x_i$ é um valor específico.

Exemplo do dado visto anteriormente.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.randint.html

*PMF* (Probability Mass Function) é a função de probabilidade para variáveis discretas. Ela é a função que mapeia um valor para a probabilidade de ocorrência. Para a distribuição uniforme, a *PMF* é constante.

*CDF* (Cumulative Distribution Function) é a função de distribuição acumulada. Ela é a função que mapeia um valor para a probabilidade de ocorrência de valores menores ou iguais a ele. Para a distribuição uniforme discreta, a *CDF* é uma função degrau.



*PPF* (Percent Point Function) é a função que mapeia uma probabilidade para o valor correspondente. Para a distribuição uniforme discreta, a *PPF* é uma função degrau.
