# Distribuições de probabilidade para variáveis contínuas


Há diversas distribuições existentes. Aqui, estudaremos algumas mais recorrentes.

## Distribuição uniforme

A distribuição uniforme é a mais simples para variáveis aleatórias contínuas, sendo utilizada para modelar a ocorrência de eventos cuja probabilidade é constante em intervalos de mesma amplitude.

Uma variável aleatória $X$ tem distribuição uniforme no intervalo $[a, b]$, se sua função densidade de probabilidade for dada por:

$$
f(x) = 
\begin{cases}
\frac{1}{b-a} & \text{ se } a \le x \le b \\
0 & \text{ caso contrário}
\end{cases}
$$

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.uniform.html

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import uniform
```

```python
uniforme = uniform.rvs(0,10,size= 100_000, random_state=42)

valores, contagem = np.unique(uniforme,return_counts=True)

#print(valores)
#print(contagem)

plt.scatter(valores, contagem,marker='o',color='red')
```

```python
import seaborn as sns

sns.histplot(uniforme,bins="auto")
plt.show()
```

*PDF* significa *Probability Density Function*. É a função que descreve a distribuição de probabilidade de uma variável aleatória contínua. A probabilidade de um evento ocorrer em um intervalo é dada pela integral da *PDF* nesse intervalo: 

$$
P(a \le X \le b) = \int_a^b f(x)dx
$$

A função densidade de probabilidade (PDF) de uma variável aleatória contínua fornece uma densidade e não uma probabilidade direta. 

Para variáveis aleatórias contínuas, a PDF $f(x)$ não fornece a probabilidade de que a variável aleatória tome exatamente o valor $x$, pois essa probabilidade é zero. Em vez disso, a PDF dá a densidade de probabilidade em torno do ponto $x$. A probabilidade de a variável aleatória cair em um intervalo específico pode ser obtida integrando a PDF sobre esse intervalo.



```python
x = np.linspace(0,10,10_000)
y = uniform.pdf(x,0,10)

plt.scatter(x,y,marker='o', color ='red')
```

```python
uniform.pdf(0.5,0,10)
```

```python
uniform.pdf(11,0,10)
```

A função de distribuição acumulada (CDF) de uma variável aleatória contínua $X$ é uma função que fornece a probabilidade de que $X$ seja menor ou igual a um valor específico $x$. Matematicamente, a CDF $F(x)$ é definida como:

$$ F(x) = P(X \leq x) $$

Para uma distribuição uniforme contínua $ U(a, b) $, a CDF é dada por:

$$ F(x) = \begin{cases} 
0 & \text{se } x < a \\
\frac{x - a}{b - a} & \text{se } a \leq x \leq b \\
1 & \text{se } x > b
\end{cases} $$

```python
x = np.linspace(0,10,10_000)
y = uniform.cdf(x,0,10)

plt.scatter(x,y,marker='o', color ='red')
```

A função de ponto percentual (PPF), também conhecida como quantile function ou inversa da CDF, fornece o valor $x$ tal que a probabilidade de $X$ ser menor ou igual a $x$ é igual a uma dada probabilidade $p$. Em outras palavras, para um dado $p$, $x = F^{-1}(p)$, onde $F^{-1}$ é a inversa da CDF.

Para uma distribuição uniforme contínua $ U(a, b) $, a PPF é dada por:

$$ F^{-1}(p) = a + p \cdot (b - a) $$

```python
uniform.ppf(0.5,0,10)

```

```python
uniform.cdf((4,6),0,10)
```

```python
sns.histplot(uniforme,bins='auto',stat="probability")
```

```python
sns.histplot(uniforme,bins=10,stat="density")
```

Quando `stat="probability"`, temos a probabilidade que um valor pertença a certa barra. As barras são normalizadas de tal forma que suas alturas somem 1. A barra mais alta significa que há maior probabilidade de o valor estar no intervalo desta barra.

Quando `stat="density"`, temos algo similar a um KDE. As barras são normalizadas de forma que a área total dos histogramas seja igual a 1.
