# Distribuições de probabilidade para variáveis contínuas

## Distribuição qui-quadrado

Uma variável aleatória contínua *X* tem distribuição qui-quadrado com $\nu$ graus de liberdade se sua função densidade de probabilidade for dada por:

$$
f(x) = 
\begin{cases}
\frac{1}{ 2^{\frac{\nu}{2}} \cdot \Gamma(\frac{\nu}{2}) } \cdot x^{\left( \frac{\nu}{2} \right) - 1} \cdot e^{-\frac{x}{2}} & x>0 \\
0 & x<0
\end{cases}
$$

em que $\Gamma (\alpha) = \int\limits_{0}^{\infty} e^{-x} \cdot x^{\alpha - 1} dx $

Grau de liberdade é, em estatística, o número de determinações independentes (dimensão da amostra) menos o número de parâmetros estatísticos a serem avaliados na população. Por exemplo, para estimar a variância de uma amostra de *N* observações, e sua estimativa depende da média da amostra, então o grau de liberdade é *N - 1*.

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2.html

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import chi2

for grau_liberdade in range (1,11):
    x = np.linspace(0,20, 10_000)
    y = chi2.pdf(x,grau_liberdade)

    plt.plot(x,y,label=f"Grau liberdade: {grau_liberdade}")


plt.ylim(0,1)
plt.xlim(0,20)
plt.legend()
```

Para valores elevados de $\nu$, a distribuição $\chi^2$ aproxima-se de uma distribuição normal padrão.

A distribuição $\chi^2$ possui diversas aplicações em inferência estatística. Sua importância se deve ao fato de podermos obter o cálculo das probabilidades e da função densidade de probabilidade acumulada para diferentes valores de *x* da variável aleatória *X*.

Lembrando do exemplo dos planos de saúde:

```python
x = np.linspace(0,20, 10_000)
y = chi2.pdf(x,4)

plt.plot(x,y,label=f"Grau liberdade: 4")
```
