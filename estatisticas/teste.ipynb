```python
import requests
import pandas as pd
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt


url = ("https://api.bcb.gov.br/dados/serie/bcdata.sgs.11/dados"
    "?formato=json&dataInicial=01/01/2024&dataFinal=31/12/2025")
dados = requests.get(url).json()


df = pd.DataFrame(dados)
df['data'] = pd.to_datetime(df['data'], dayfirst=True)
df['valor'] = df['valor'].astype(float)



print(df.head())
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax,plt.plot(df['data'], df['valor'])

ax.yaxis.set_major_formatter(mtick.PercentFormatter(1.0))

#for i, valor in enumerate(df['valor']):
#     ax.text(i,valor, f'% {valor:.2f} MI'.replace('.', ','),ha='center',va='bottom',rotation=10,fontsize=10)

ax.set_xlabel('Data')
ax.set_ylabel('%')

plt.show()
```
