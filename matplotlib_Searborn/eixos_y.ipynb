```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
precipitacao_media = [200, 180, 160, 100, 60, 30, 30, 35, 60, 100, 160, 200]
temperatura_media = [30, 29, 27, 25, 23, 22, 22, 23, 24, 25, 26, 28]
```

```python
fig, ax1 = plt.subplots()

cor_temperatura='tab:red'
cor_precipitacao ='tab:blue'

ax1.bar(meses,precipitacao_media,color=cor_precipitacao,label='Precipitao')
ax1.set_ylabel("precipitao media",color=cor_precipitacao)
ax1.tick_params(axis='y',labelcolor=cor_precipitacao)
ax1.set_ylim(0,260)
ax2 = ax1.twinx()
ax2.plot(meses, temperatura_media,color=cor_temperatura,label="temperatura Mdia")
ax2.set_ylabel("precipitao media",color=cor_temperatura)
ax2.tick_params(axis='y',labelcolor=cor_temperatura)
ax2.set_ylim(20,30)

ax1.set_title("Temperatura e precipitao")
fig.legend(bbox_to_anchor=(1,1),bbox_transform=ax1.transAxes)


plt.show()
```
