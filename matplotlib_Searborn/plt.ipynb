```python
import pandas as pd

arquivo_acoes = 'top_10_ibov_mai_ago_2023.csv'

df_top10_ibov = pd.read_csv(arquivo_acoes)

df_top10_ibov.head()


```

```python
df_top10_ibov.info()
```

```python
df_top10_ibov['Date'] = pd.to_datetime(df_top10_ibov['Date'], format ='%Y-%m-%d')
df_top10_ibov.info()
```

```python
df_top10_ibov = df_top10_ibov.set_index('Date', drop= True)
df_top10_ibov.head()
```

```python
df_top10_ibov.describe()
```

```python
df_top10_ibov.plot()
```

```python
import matplotlib.pyplot as plt

plt.style.use('bmh')

df_top10_ibov.plot(figsize=(12,9))

plt.show()
```

```python
fig,ax = plt.subplots(figsize=(12,9))

df_top10_ibov.plot(ax=ax)
ax.set_title("Top 10 Aes")
plt.show()
```

```python
df_top10_ibov.plot(figsize=(10,12), title='aes', subplots=True )
```

```python
df_top10_ibov.plot(figsize=(10,12), title='aes', subplots=True,layout=(5,2) )
```

```python
df_top10_ibov.index
```

```python
fig, ax = plt.subplots()

ax.plot(df_top10_ibov.index, df_top10_ibov['BBDC4'],label='bbdc4')
ax.plot(df_top10_ibov.index, df_top10_ibov['ITUB4'],label='itub4')
ax.plot(df_top10_ibov.index, df_top10_ibov['BBAS3'], label='bbas3')

ax.set_title("Comparando Aes de Bancos")
ax.set_ylabel("Valor em R$")
ax.legend()


plt.show()
```

```python
variacao = df_top10_ibov.pct_change()
variacao.head()
```

```python
fig, ax = plt.subplots()

ax.plot(variacao.index,variacao['ITUB4'],label='itub4')
ax.plot(variacao.index,variacao['BBDC4'],label='bbdc4')
ax.plot(variacao.index,variacao['BBAS3'], label='bbas3')

ax.set_title("Variao Aes de Bancos")
ax.set_ylabel("variao em %")
ax.legend()


plt.show()
```

```python
fig, ax = plt.subplots(nrows=3,ncols=1,figsize=(10,10))

bancos=["BBDC4","ITUB3",'BBAS3']

ax[0].plot(variacao.index,variacao['ITUB4'],label='itub4',color='C0')
ax[0].legend()
ax[1].plot(variacao.index,variacao['BBDC4'],label='bbdc4',color='C1')
ax[1].legend()
ax[2].plot(variacao.index,variacao['BBAS3'], label='bbas3',color='C2')
ax[2].legend()

plt.show()
```

```python
fig, axs = plt.subplots(nrows=3,ncols=1,figsize=(10,10))

bancos=["BBDC4","ITUB4",'BBAS3']

for i,banco in enumerate (bancos):

    axs[i].plot(variacao.index, variacao[banco],label=banco,color=f'C{i}')
    axs[i].legend()
    axs[i].set_ylabel("variao %")

    
fig.suptitle("Variao Aes de Bancos")

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3,figsize=(10,6))

axs[0].scatter('BBDC4', 'BBAS3', data=df_top10_ibov)
axs[0].set_xlabel('BBDC4')
axs[0].set_ylabel('BBAS3')
axs[0].set_title("Bradesco X Banco do Brasil")


axs[1].scatter('BBDC4', 'ITUB4', data=df_top10_ibov)
axs[1].set_xlabel('BBDC4')
axs[1].set_ylabel('ITUB')
axs[1].set_title("Bradesco X Itau")


axs[2].scatter('BBAS3', 'ITUB4', data=df_top10_ibov)
axs[2].set_xlabel('BBAS3')
axs[2].set_ylabel('ITUB')
axs[2].set_title("Banco do Brasil X Itau")


fig.suptitle("Analisando correlo dos bancos")
plt.show()
```

```python
from itertools import combinations


pares = list(combinations(bancos,2))

print (pares)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=3,figsize=(10,6))

for i , par in enumerate(pares):
    axs[i].scatter(par[0],par[1],data=df_top10_ibov)
    axs[i].set_xlabel(par[0])
    axs[i].set_ylabel(par[1])
    axs[i].set_title(f'{par[0]} X {par[1]}')

fig.suptitle("Analisando correleo dos bancos")
plt.show()
```
