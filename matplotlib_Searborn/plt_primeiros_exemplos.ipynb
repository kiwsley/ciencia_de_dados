```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
```

```python
plt.plot(meses,vendas,marker='o')
plt.xlabel('Meses')
plt.ylabel('Vendas R$M')
plt.grid(True)
plt.show()
```

```python
cidades = ['So Paulo', 'Rio de Janeiro', 'Salvador']
temperaturas = [20, 25, 27]

plt.bar(cidades,temperaturas,color="lightgreen")
plt.ylabel("Temperatura Mdia em C")
plt.title("Comparando temperatura mdia")
plt.show()
```

```python
cidades = ['So Paulo', 'Rio de Janeiro', 'Salvador']
temperaturas = [20, 25, 27]

barra=plt.bar(cidades,temperaturas,color="lightgreen")
plt.bar_label(barra,labels=temperaturas,padding=1)
plt.ylabel("Temperatura Mdia em C")
plt.title("Comparando temperatura mdia")
plt.show()
```
