```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]  # valores em porcentagem


plt.title('Distribuição de Categorias de Produtos em Porcentagem')
plt.pie(quantidades, labels=categorias)
plt.show()
```

```python
import matplotlib.pyplot as plt

categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]  # valores em porcentagem
destaque = [0,0,0,0.1]


plt.title('Distribuição de Categorias de Produtos em Porcentagem')
plt.pie(quantidades, labels=categorias,autopct='%.1f',explode=destaque)
plt.show()
```
