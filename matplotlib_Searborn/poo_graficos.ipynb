```python
import matplotlib.pyplot as plt

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

fig, ax = plt.subplots()

ax.plot(meses, vendas)
ax.set_xlabel("Mês")
ax.set_ylabel("Vendas em R$")
ax.set_title("Evolução de vendas ao longo do Ano")
plt.show()
```

```python
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
quantidades = [23, 45, 17, 30, 37]

fig, ax = plt.subplots()

ax.bar(produtos,quantidades)
ax.set_xlabel("Produtos")
ax.set_ylabel("Quantidade de vendas")
ax.set_title("quantidades de vendas por produto")
plt.show()
```

```python
categorias = ['Eletrônicos', 'Roupas', 'Alimentos', 'Livros']
quantidades = [45, 30, 15, 10]
destaque = [0.1, 0, 0, 0]  # destacar a primeira fatia

fig, ax = plt.subplots()

ax.pie(quantidades,labels=categorias,explode=destaque)

plt.show()
```

```python
anos = [2018, 2019, 2020, 2021]
vendas_online = [5, 10, 15, 20]
vendas_loja = [10, 8, 6, 4]

fig, ax = plt.subplots()

ax.bar(anos, vendas_online, label='Online')
ax.bar(anos, vendas_loja, bottom=vendas_online, label='Loja')

ax.set_xlabel('Ano')
ax.set_ylabel('Vendas (em milhões de reais)')
ax.set_title('Vendas online vs loja')

ax.set_xticks(anos)

ax.legend()

plt.show()
```
