```python
import matplotlib.pyplot as plt
import numpy as np
from cycler import cycler
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas, marker='o',linestyle='--')
plt.annotate('Menor Valor', color = 'red',xy=(2,78),xytext=(3,80),arrowprops= dict(facecolor='red'))
plt.xlabel('MÃªs')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
marcadores = np.where(vendas == np.min(vendas),'x','')

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]

plt.plot(meses, vendas)
for posicao , marcador in enumerate(marcadores):
    if marcador =='x':
        cor='red'
        tamanho=80
    else:
        cor='blue'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,color = cor, s=tamanho)

plt.annotate('Menor Valor', color = 'red',xy=(2,78),xytext=(2.2,78))
plt.xlabel('MÃªs')
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```

```python
marcadores = np.where(vendas == np.min(vendas),'x','')

meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
vendas = [85, 90, 78, 92, 88, 96, 94, 98, 100, 102, 105, 108]
#paleta de cores
cores = plt.get_cmap('Pastel2').colors
ciclo_cores=cycler('color',cores)
plt.rc('axes',prop_cycle=ciclo_cores)

#grafico
plt.plot(meses, vendas)
#marcadores
for posicao , marcador in enumerate(marcadores):
    if marcador =='x':
        cor='red'
        tamanho=80
    else:
        cor='black'
        tamanho=0
    plt.scatter(meses[posicao],vendas[posicao],marker=marcador,color = cor, s=tamanho)
    plt.annotate(vendas[posicao], color = cor,xy=(meses[posicao],vendas[posicao]),xytext=(-5,7),textcoords='offset points')

#destaque da regiÃ£o
plt.axvspan(6,11,alpha =0.3,color=cores[1])
plt.text(6,80,"RecuperaÃ§Ã£o de crescimento",color ='green',weight='bold',fontsize=9)

plt.box(False)
plt.tick_params(axis='x',length=0)
plt.yticks([])
plt.ylabel('Vendas (em milhares)')
plt.title('Crescimento das Vendas ao Longo do Ano')
plt.show()
```
